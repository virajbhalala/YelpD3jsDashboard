```python
data2 =pd.read_csv("./AppleStore.csv")
```

```python
data2
```

```python
newdata = pd.DataFrame()
newdata["average_user_rating"] = data2.groupby('prime_genre').user_rating.mean()
newdata["average_rating_counts"] = data2.groupby('prime_genre').rating_count_tot.mean()
newdata["price"] = data2.groupby('prime_genre').price.mean()
newdata["average_supported_devices"] = data2.groupby('prime_genre')["sup_devices.num"].mean()
newdata["average_supported_languages"] = data2.groupby('prime_genre')["lang.num"].mean()
```

```python
newdata.to_csv("./bar2.csv")
```
